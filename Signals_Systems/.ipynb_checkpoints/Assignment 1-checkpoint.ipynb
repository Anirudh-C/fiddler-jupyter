## Problem Statement
$x(t) = 1, \vert t  - 10k \vert \le 2, k \in \mathbb{Z}$

```python
import matplotlib.pyplot as plt
import math
```

```python
import scipy
from scipy.integrate import quad
import numpy as np
```

Consider the Fourier Series expansion of the above function

$x(t) = \sum_{k = - \infty}^{\infty} a_k e^{jk\omega_0 t}$

where $\omega_0 = \frac{2 \pi}{T}$

$a_k = \frac{1}{T} \int_{T} x(t) e^{-j k \omega_0 t} dt$

For the given function,

$T = 10$

So consider the integral from $0$ to $T$

$a_k = \frac{1}{10} \int_0^{10} x(t) e^{-j k \omega_0 t} dt$

$a_k = \frac{1}{10} \left( \int_0^2 e^{-j k \omega_0 t}dt + \int_8^{10} e^{- j k \omega_0 t}dt \right)$ 

$a_k = \frac{1}{5} \int_0^2 e^{- j k \omega_0 t}dt$

For k = 0,

$a_0 = \frac{2}{5}$

Otherwise,

$a_k = \frac{\sin \left( \frac{2k\pi}{5} \right)}{k\pi}$

```python
def x(t):
    if (t <= 2 and t>= 0) or (t <= 10 and t>= 8):
        return 1
    return 0
plt.plot(np.linspace(0,10),np.vectorize(x)(np.linspace(0,10)),label="x(t)")
plt.legend()
plt.xlabel("t")
plt.ylabel("x(t)")
plt.show()
```

```python
def complex_integral(func, a, b, **kwargs):
    def real_func(x):
        return scipy.real(func(x))
    def imag_func(x):
        return scipy.imag(func(x))
    real_integral = quad(real_func, a, b, **kwargs)
    imag_integral = quad(imag_func, a, b, **kwargs)
    return real_integral[0]
```

```python
def a(k):
    """Computes the Fourier Coefficients of x(t)"""
    return (1/10)*complex_integral(lambda x: (scipy.exp(1j * (scipy.pi/5) * k * x*(-1))), -2, 2)
```

```python
k = np.arange(-20, 20)
plt.scatter(k,np.vectorize(a)(k),label="a(k)")
plt.plot(k,np.vectorize(a)(k),label="Approx a(k)",color="orange")
plt.xlabel("k")
plt.ylabel("a(k)")
plt.legend()
plt.show()
```

```python
def fourier_series(k,t):
    real = 0
    imag = 0
    for i in range(k):
        real += a(k) * np.cos(k*(np.pi/5)*t)
        imag += a(k) * np.sin(k*(np.pi/5)*t)
    return (real, imag)
```

The following are the plots for the real and imaginary parts of $x_2(t)$

```python
t = np.linspace(0,10)
x2_real = np.zeros(t.size)
x2_imag = np.zeros(t.size)
for i in range(t.size):
    x2_real[i] = fourier_series(2,t[i])[0]
    x2_imag[i] = fourier_series(2,t[i])[1]
plt.plot(t,x2_real,label="Re(x2(t))")
plt.plot(t,x2_imag,label="Im(x2(t))")
plt.xlabel("t")
plt.legend()
plt.show()
```

The following are the plots for the real and imaginary parts of $x_{10}(t)$

```python
x10_real = np.zeros(t.size)
x10_imag = np.zeros(t.size)
for i in range(t.size):
    x10_real[i] = fourier_series(10,t[i])[0]
    x10_imag[i] = fourier_series(10,t[i])[1]
plt.plot(t,x2_real,label="Re(x10(t))")
plt.plot(t,x2_imag,label="Im(x10(t))")
plt.xlabel("t")
plt.legend()
plt.show()
```

## Error
Given $x(t)$ and $x_k(t)$, we define the error as

$e_k(t) = \vert\vert x(t) - x_k(t) \vert\vert$

```python
def e(k,t):
    norm = ((x(t) - fourier_series(k,t)[0])**2 + (fourier_series(k,t)[1])**2)**(0.5)
    return norm
```

The error for $x_2(t)$, $e_2(t)$

```python
e2 = np.zeros(t.size)
for i in range(t.size):
    e2[i] = e(2,t[i])
plt.plot(t,e2,label="e2(t)")
plt.legend()
plt.xlabel("t")
plt.ylabel("e2(t)")
plt.show()
```

The error for $x_2(t)$, $e_2(t)$

```python
e10 = np.zeros(t.size)
for i in range(t.size):
    e10[i] = e(10,t[i])
plt.plot(t,e10,label="e10(t)")
plt.legend()
plt.xlabel("t")
plt.ylabel("e10(t)")
plt.show()
```
