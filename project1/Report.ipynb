# Finite Difference Approximation

Consider a function $u$ defined on a domain $[a,b]$.

Create a partition on this domain of $N$ points $x_{i}$ such that
$$x_{i} = a + i \triangle x$$
where,
$\triangle x = \frac{b-a}{N-1}$, is the size of the partition.

## First Differential Approximation

Let $u_{i}$ represent the values of $u$ at the partition points ${x_{i}}$

That is,
$$u_{i} = u(x_{i})$$
By Taylor's expansion,
$$u_{i+1} = u(x_{i}) + u'(x_{i})\triangle x + \frac{u''(x_{i})\triangle x^{2}}{2} + \cdot \cdot \cdot$$
Thus,
$$u_{i+1} - u_{i} = u'(x_{i})\triangle x + O(\triangle x^{2})$$
Dividing by $\triangle x$ on both sides we get
$$\frac{u_{i+1} - u_{i}}{\triangle x} = u'(x_{i}) + O(\triangle x)$$
This is the forward difference approximation for $u'$
    
Similarly, a backward difference approximation can be made for $u'$ using the Taylor's expansion for $u_{i-1}$
$$u_{i-1} = u(x_{i}) - u'(x_{i})\triangle x + \frac{u''(x_{i})\triangle x^{2}}{2} + \cdot \cdot \cdot$$
Therefore,
$$u_{i} - u_{i-1} = u'(x_{i})\triangle x + O(\triangle x^{2})$$
Dividing by $\triangle x$ on both sides we get
$$\frac{u_{i} - u_{i-1}}{\triangle x} = u'(x_{i}) + O(\triangle x)$$  
Subtracting the Taylor expansions for $u_{i+1}$ and $u_{i-1}$, we get
$$u_{i+1} - u_{i-1} = 2 u'(x)\triangle x + O(\triangle x^{2})$$
Dividing by $2\triangle x$ on both sides we get
$$\frac{u_{i+1} - u_{i-1}}{2\triangle x} = u'(x_{i}) + O(\triangle x)$$

## Second Differential Approximation

Using the Taylor expansions of $u_{i+1}$ and $u_{i-1}$ an approximation for the second differential $u''$ of a function $u$ can be found

Adding the Taylor expansions for $u_{i+1}$ and $u_{i-1}$, we get
$$u_{i+1} + u_{i-1} = 2 u(x) + u''(x)\triangle x^{2} + O(\triangle x^{4})$$
Thus,
$$u_{i+1} + u_{i-1} - 2 u_{i} = u''(x)\triangle x^{2} + O(\triangle x^{4})$$
Dividing by $\triangle x^{2}$ on both sides we get
$$\frac{u_{i+1} -2u_{i} + u_{i-1}}{\triangle x^{2}} = u''(x_{i}) + O(\triangle x^{2})$$

# 1-Dimensional Boundary Value Problem

Consider an equation of the form
$$-u'' = f$$
defined over an interval $[a,b]$ and where $f = f(x)$ is a real-valued function

The boundary conditions for this equation are given as
$$u(a) = \alpha$$
$$u(b) = \beta$$

## Forming the Approximation
Let $U_{i}$ be the approximation for the function $u$ using the partition scheme defined above,

By using the Finite Difference Approximation for $u''$, we get
$$U_{0} = \alpha$$
$$-\frac{U_{i+1} -2U_{i} + U_{i-1}}{\triangle x^{2}} = f_{i}, \qquad i = 1,2,\ldots,N-1$$
$$U_{N-1} = \beta$$
These equations can be re-written as,
$$U_{0} = \alpha$$
$$-U_{i+1} + 2U_{i} - U_{i-1} = \triangle x^{2}f_{i}, \qquad i = 1,2,\ldots,N-1$$
$$U_{N-1} = \beta$$
This system of $N$ equations in the vector $U$ can be represented as a matrix-vector equation
$$Ax = b$$
Where,
$$A = \left[\begin{array}{A}
1 & 0 & 0 & 0 & \ldots & 0\\
-1 & 2 & -1 & 0 & \ldots & 0 \\
0 & -1 & 2 & -1 & \ldots & 0\\
\vdots & & & & & \vdots \\
0 & \ldots & 0 & -1 & 2 & -1 \\
0 & \ldots & \ldots & 0 & 0 & 1
\end{array}\right], \qquad
x = \left[\begin{array}{x}
U_{0}\\
U_{1}\\
\vdots\\
U_{N-1}
\end{array}\right], \qquad
b = \left[\begin{array}{b}
\alpha\\
\triangle x^{2}f(x_{1})\\
\triangle x^{2}f(x_{2})\\
\vdots\\
\triangle x^{2}f_(x_{N-2})\\
\beta
\end{array}\right]$$

This is a case of the Thomas tri-diagonal Algorithm

## Thomas Tri-Diagonal Algorithm
The general form of an $N \times N$ matrix is 
$$
A = \left[\begin{array}{generalA}
a_0 & c_0 & 0 & 0 & \ldots & 0\\
b_0 & a_1 & c_1 & 0 & \ldots & 0 \\
0 & b_1 & a_2 & c_2 & \ldots & 0\\
\vdots & & & & & \vdots \\
0 & \ldots & 0 & b_{N-3} & a_{N-2} & c_{N-2} \\
0 & \ldots & \ldots & 0 & b_{N-2} & a_{N-1}
\end{array}\right]
$$
While LU decomposing the above matrix, most of the elements of the matrices $L$ and $U$ will be $0$
Thus, the form of the matrices $L$ and $U$ is 
$$
L = \left[\begin{array}{L}
\alpha_0 & 0 & 0 & 0 & \ldots & 0\\
b_0 & \alpha_1 & 0 & 0 & \ldots & 0 \\
0 & b_1 & \alpha_2 & 0 & \ldots & 0\\
\vdots & & & & & \vdots \\
0 & \ldots & 0 & b_{N-3} & \alpha_{N-2} & 0 \\
0 & \ldots & \ldots & 0 & b_{N-2} & \alpha_{N-1}
\end{array}\right], \qquad
U = \left[\begin{array}{U}
1 & \gamma_0 & 0 & 0 & \ldots & 0\\
0 & 1 & \gamma_1 & 0 & \ldots & 0 \\
0 & 0 & 1 & \gamma_2 & \ldots & 0\\
\vdots & & & & & \vdots \\
0 & \ldots & 0 & 0 & 1 & \gamma_{N-2} \\
0 & \ldots & \ldots & 0 & 0 & 1
\end{array}\right]
$$
By multiplying the matrices $L$ and $U$ and equating to $A$ we get,
$$a_{0} = \alpha_{0}, \qquad \alpha_{0}\gamma_{0} = c_{0}$$
$$a_{i} = \alpha_{i} + b_{i-1}\gamma_{i-1}, \qquad i = 1,\ldots,N-1$$
$$\alpha_{i}\gamma_{i} = c_{i}, \qquad i = 1,\ldots,N-2$$
Solving this system for ${\alpha_{i}}$ and ${\gamma_{i}}$, we get
$$\alpha_{0} = a_{0}, \qquad \gamma_{0} = \frac{c_{0}}{\alpha_{0}}$$
$$\alpha_{i} = a_{i} - b_{i-1}\gamma_{i-1}; \qquad \gamma_{i} = \frac{c_{i}}{\alpha_{i}}, \qquad i = 1,\ldots,N-2$$
$$\alpha_{N-1} = a_{N-1} - b_{N-2}\gamma_{N-2}$$
To solve $LUx = f$ we assume $Ux = z$ and $Lz = f$, to get
$$z_{0} = \frac{f_{0}}{\alpha_{0}}$$
$$z_{i} = \frac{f_{i} - b_{i-1}z_{i-1}}{\alpha_{i}}, \qquad i = 1,\ldots,N-1$$
$$x_{N-1} = z_{N-1}$$
$$x_{i} = z_{i} - \gamma_{i}x_{i+1}$$

## Solution
Applying the Thomas algorithm to the equation formed by the 1-D BVP, we get
$$z_{0} = \frac{f_{0}}{\alpha_{0}}$$
$$z_{i} = \frac{f_{i} - b_{i-1}z_{i-1}}{\alpha_{i}}, \qquad i = 1,\ldots,N-1$$
$$U_{N-1} = z_{N-1}$$
$$U_{i} = z_{i} - \gamma_{i}U_{i+1}$$
where,
$$f_{0} = \alpha$$
$$f_{i} = f(x_{i}), \qquad i = 1,\ldots,N-2$$
$$f_{N-1} = \beta$$

# Implementation

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
from matplotlib import pyplot as plt
import math
```

## LU_decompose
The method takes three diagonals of a tri-diagonal matrix P, L and U and returns the pair of diagonals alpha-vector and gamma-vector, where
$$P = [a_{i}]$$
$$L = [b_{i}]$$
$$U = [c_{i}]$$
$$alphavector = [\alpha_{i}]$$
$$gammavector = [\gamma_{i}]$$

```python
def LU_decompose(P,L,U):
    """
    This method takes 3 diagonals, P, L and U in that order, where
    P is the principal diagonal, L is the lower diagonal and U is the upper diagonal
    of a tri-diagonal matrix and returns 2 diagonals packed together as
    alpha_vector and gamma_vector, where alpha_vector is the principal diagonal of
    the matrix L and gamma_vector is the upper diagonal of the matrix U in the
    LU-decomposition of A
    """
    N = len(P)
    #Initialisation of output arrays: O(1)
    alpha_vector = np.zeros(N)
    gamma_vector = np.zeros(N-1)
    #alpha and gamma calculation
    ##Initialisation: O(1)
    alpha_vector[0] = P[0]
    gamma_vector[0] = U[0]/alpha_vector[0]
    ##Recursive calculation: O(n)
    for i in range(1,N-1):
        alpha_vector[i] = P[i] - L[i-1] * gamma_vector[i-1]
        gamma_vector[i] = U[i]/alpha_vector[i]
    ##Final Value: O(1)
    alpha_vector[N-1] = P[N-1] - L[N-2]*gamma_vector[N-2]
    return alpha_vector,gamma_vector
```

## LU_solve
This method takes the alpha_vector,gamma_vector,f_vector and L and returns the solution to $LUx=f$
The vector f_vector is $f(x)$ where $x$ is the vector of $N$ points in the domain $[a,b]$

```python
def LU_solve(alpha_vector,gamma_vector,f_vector,L):
    """
    This method takes the principal diagonal of the lower triangular matrix, the upper
    diagonal of the upper triangular matrix, the function vector at all points in the discretised
    domain and the lower diagonal of the matrix A and returns the solution to LUx = f
    """
    N = len(alpha_vector)
    #Initialisation
    z = np.ones(N)
    U = np.ones(N)
    #Solving Lz = f for z
    z[0] = f_vector[0]/alpha_vector[0]
    for i in range(1,N):
        z[i] = (f_vector[i] - L[i-1]*z[i-1])/alpha_vector[i]
    #Solving Ux = z for x
    U[N-1] = z[N-1]
    for i in range(N-2,-1,-1):
        U[i] = z[i] - gamma_vector[i]*U[i+1]
    return U
```

## formMatrix
This method takes the number of points N in the domain and the domain d to return the matrix characterized by the 1-D BVP 

```python
def formMatrix(N,d):
    """
    This method takes an integer N and returns the three diagonals of the matrix formed by
    discretising the first degree Poisson equation -u'' = f, with N-1 intervals on the
    domain
    """
    a,b = d
    P = [2]*N
    L = [-1]*(N-1)
    U = [-1]*(N-1)
    #Principal diagonal calculation: O(1)
    P[0] = 1
    P[N-1] = 1
    #Lower diagonal calculation: O(1)
    L[N-2] = 0
    #Upper diagonal calculation: O(1)
    U[0] = 0
    return P,L,U
```

## Driver
This method takes the input N, function f, and boundary condition values, alpha and beta to return the solution of the 1-D BVP.

```python
def driver(N,f,d,alpha,beta):
    """
    This method takes the inputs N(number of points taken in the domain), f the given function,
    the domain d, and the boundary values to give an N-vector of the approximate
    solution to the first degree Poisson equation.
    """
    P,L,U = formMatrix(N,d)
    alpha_vector, gamma_vector = LU_decompose(P,L,U)
    x = np.linspace(d[0],d[1],N)
    dx = (d[1] - d[0])/(N-1)
    g = lambda x: dx*dx*f(x)
    vfunc = np.vectorize(g)
    f_vector = vfunc(x)
    f_vector[0] = alpha
    f_vector[N-1] = beta
    solution = LU_solve(alpha_vector,gamma_vector,f_vector,L)
    return solution
```

# Test Cases

## Test Case 1:
$$f = 1$$
$$u = -x^{2}/2 + x/2$$

```python
def test1():
    f = lambda x: 1
    u = lambda x: x/2 - x*x/2
    d = (0,1)
    alpha = 0
    beta = 0
    N = 20
    x = np.linspace(d[0],d[1],N)
    vfunc = np.vectorize(u)
    exact = vfunc(x)
    approx = driver(N,f,d,alpha,beta)
    plt.plot(x,exact,x,approx,'o')
    plt.xlabel('x')
    plt.ylabel('u')
    plt.grid(True)
    plt.legend(("Exact","Approximation"))
test1()
```

# Error Analysis

## Test Case 1:
$$f = 1$$
$$u = -x^{2}/2 + x/2$$

```python
N = [20,40,80,160,320]
def errorplot1():
    f = lambda x: 1
    u = lambda x: x/2 - x*x/2
    d = (0,1)
    alpha = 0
    beta = 0
    error = []
    for i in range(5):
        x = np.linspace(d[0],d[1],N[i])
        vfunc = np.vectorize(u)
        exact = vfunc(x)
        approx = driver(N[i],f,d,alpha,beta)
        error.append(np.abs(approx-exact).max())
    plt.loglog(N,error,'o')
    plt.xlabel('x')
    plt.ylabel('error')
    plt.grid(True)
errorplot1()
```
